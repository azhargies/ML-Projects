```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from tensorflow import keras
from tensorflow.keras import Sequential
```

```python
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
X = data.drop(["Outcome"], axis = 1)
Y = data["Outcome"]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x_scaled, Y, test_size = 0.2, random_state = 42)
```

```python
x_scaled[:2]
```

```python
model = Sequential(
    [keras.layers.Dense(units = 1, input_shape = (8,), activation = "sigmoid")]
)

model.summary()

model.compile(keras.optimizers.Adam(
    learning_rate=0.01), loss = "binary_crossentropy", metrics = ["accuracy"])

model.fit(xtrain, ytrain, epochs = 1000)
```

```python
model_new = Sequential()

model_new.add(keras.layers.Dense(128, input_shape = (8,), activation = "relu"))
model_new.add(keras.layers.Dropout(0.2))
model_new.add(keras.layers.Dense(128, input_shape = (8,), activation = "relu"))
model_new.add(keras.layers.Dropout(0.2))
model_new.add(keras.layers.Dense(units = 1, activation = "sigmoid"))

model_new.summary()

model_new.compile(keras.optimizers.Adam(
    learning_rate=0.01), loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model_new.fit(xtrain, ytrain, epochs = 1000)
```

```python
model_new.evaluate(xtest,ytest)
```
